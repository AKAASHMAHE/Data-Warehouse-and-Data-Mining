```python
import pandas as pd

df = pd.read_excel('Employee Dataset.xlsx')
```

```python
print("Dimensions (rows, columns):", df.shape)
print("\nStructure of dataset:")
print(df.info())
print("\nAttribute names:")
print(df.columns.tolist())
```

```python
for col in df.columns:
    print(f"Unique values in '{col}':")
    print(df[col].unique())
    print()
```

```python

print("First 5 Records:")
print(df.head())


print("\nLast 5 Records:")
print(df.tail())


print("\nName, Designation, Salary of first 10 records:")
print(df.loc[:9, ['Name', 'Designation', 'Salary']])


print("\nNames of all records:")
print(df['Name'])


print("\nAll records:")
print(df)
```
