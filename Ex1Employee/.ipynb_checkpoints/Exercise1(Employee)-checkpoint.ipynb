```python
import pandas as pd
import matplotlib.pyplot as plt
#Akaash M K 23MIC0012
df = pd.read_excel('Employee Dataset.xlsx')
```

```python
print("Dimensions (rows, columns):", df.shape)
print("\nStructure of dataset:")
print(df.info())
#Akaash M K 23MIC0012
print("\nAttribute names:")
print(df.columns.tolist())
```

```python
for col in df.columns:
    print(f"Unique values in '{col}':")
    #Akaash M K 23MIC0012
    print(df[col].unique())
    print()
```

```python
print("First 5 Records:")
print(df.head())

print("\nLast 5 Records:")
print(df.tail())
 #Akaash M K 23MIC0012
print("\nName, Designation, Salary of First 10 Records:")
print(df.loc[:9, ['Name', 'Designation', 'Salary']])

print("\nName of All Records:")
print(df['Name'])
 #Akaash M K 23MIC0012
print("\nAll Records:")
print(df)
```

```python
print("Mean of numerical variables:")
print(df.mean(numeric_only=True))

print("\nMedian of numerical variables:")
print(df.median(numeric_only=True))

print("\nMode of numerical variables:")
print(df.mode(numeric_only=True).iloc[0])
 #Akaash M K 23MIC0012
print("\nVariance of numerical variables:")
print(df.var(numeric_only=True))

print("\nCovariance matrix:")
print(df.cov(numeric_only=True))

print("\nCorrelation between Salary and Experience:")
print(df['Salary'].corr(df['Experience']))
```

```python
designation_counts = df['Designation'].value_counts()
plt.figure(figsize=(6,6))
 #Akaash M K 23MIC0012
plt.pie(designation_counts, labels=designation_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Designation Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df['Salary'], bins=10, color='skyblue', edgecolor='black')
plt.title('Salary Distribution')
 #Akaash M K 23MIC0012
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df['Experience'], df['Salary'], color='green')
plt.title('Salary vs Experience')
 #Akaash M K 23MIC0012
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.grid(True)
plt.show()
```
