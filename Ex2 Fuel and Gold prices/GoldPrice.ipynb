```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
#Akaash M K 23MIC0012
```

```python
df = pd.read_excel('Gold Price 1999-2018 with price per gram.xlsx')
df['Date'] = pd.to_datetime(df['Date'])
#Akaash M K 23MIC0012
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
```

```python
yearly_mean = df.groupby('Year')['Price'].mean()
#Akaash M K 23MIC0012
plt.figure(figsize=(10,5))
plt.plot(yearly_mean.index, yearly_mean.values, marker='o', linestyle='-', color='gold')
plt.title('Yearly Average Gold Price (1999-2007)')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.grid(True)
plt.show()
```

```python
monthly_avg = df.groupby(['Year', 'Month'])['Price'].mean().unstack()
#Akaash M K 23MIC0012
monthly_avg.plot(kind='bar', stacked=True, figsize=(15,7), colormap='viridis')
plt.title('Stacked Bar Chart: Monthly Average Gold Price per Year')
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Akaash M K 23MIC0012
plt.figure(figsize=(12,6))
sns.boxplot(x='Year', y='Price', data=df, hue='Year', palette='coolwarm', legend=False)
plt.title('Boxplot: Gold Price Distribution per Year')
plt.xticks(rotation=45)
plt.show()
```

```python
pivot = df.groupby(['Year','Month'])['Price'].mean().unstack()
plt.figure(figsize=(12,6))
#Akaash M K 23MIC0012
sns.heatmap(pivot, annot=True, fmt=".1f", cmap='YlOrBr')
plt.title('Heatmap: Monthly Average Gold Price (1999-2007)')
plt.xlabel('Month')
plt.ylabel('Year')
plt.show()
```

```python

plt.figure(figsize=(10,6))
sorted_yearly = yearly_mean.sort_values()
#Akaash M K 23MIC0012
plt.barh(sorted_yearly.index, sorted_yearly.values, color=plt.cm.viridis(range(len(sorted_yearly))))
plt.xlabel('Average Price')
plt.ylabel('Year')
plt.title('Ranking of Average Yearly Gold Price (19992007)')
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.show()
```
