```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
from scipy.signal import find_peaks
from pandas.plotting import autocorrelation_plot
```

```python
df = pd.read_excel('Petrol price of Hyd 2012-2020.xlsx')
# Akaash M K 23MIC0012
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values('date')
```

```python
plt.figure(figsize=(12,6))
plt.step(df['date'], df['rate'], where='post', color='blue')
plt.title('Step Plot: Petrol Price Changes Over Time')
# Akaash M K 23MIC0012
plt.xlabel('Date')
plt.ylabel('Rate')
plt.grid(True)
plt.show()
```

```python
df['Rolling_30'] = df['rate'].rolling(window=30, min_periods=1).mean()

plt.figure(figsize=(10,6))
plt.plot(df['date'], df['rate'], label='Daily Rate', alpha=0.4)
# Akaash M K 23MIC0012
plt.plot(df['date'], df['Rolling_30'], color='red', linewidth=2, label='30-Day Rolling Avg')
plt.title('30-Day Rolling Average Petrol Price')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['rate'], kde=True, color='green', bins=15)
plt.title('Distribution of Petrol Prices')
# Akaash M K 23MIC0012
plt.xlabel('Rate')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df['date'], df['rate'], label='Price', color='green')
# Akaash M K 23MIC0012
peaks, _ = find_peaks(df['rate'], distance=10)
plt.plot(df['date'].iloc[peaks], df['rate'].iloc[peaks], "ro", label='Peaks')
plt.title('Petrol Price Over Time with Peaks Highlighted')
plt.xlabel('Date')
plt.ylabel('Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
autocorrelation_plot(df['rate'])
# Akaash M K 23MIC0012
plt.title('Autocorrelation Plot: Petrol Prices')
plt.show()
```
